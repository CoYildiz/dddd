## Organize the Dataset


```python
import os
import shutil

source_data_path = '/content/smap_data/data/data'
source_labels_path = '/content/smap_data/labeled_anomalies.csv'
destination_path = '/content/data/SMAP_MSL'


os.makedirs(destination_path, exist_ok=True)

if os.path.exists(source_data_path):
    if os.path.isdir(source_data_path):
        for item in os.listdir(source_data_path):
            s = os.path.join(source_data_path, item)
            d = os.path.join(destination_path, item)
            if os.path.isdir(s):
                shutil.copytree(s, d, dirs_exist_ok=True)
            else:
                shutil.copy2(s, d)
    else:
        print(f"Source data path is not a directory: {source_data_path}")
else:
    print(f"Source data path does not exist: {source_data_path}")


if os.path.exists(source_labels_path):
    shutil.copy2(source_labels_path, destination_path)
else:
    print(f"Source labels file does not exist: {source_labels_path}")

print(f"Relevant files from /content/smap_data moved to {destination_path}")
```

```python
import matplotlib.pyplot as plt
import scienceplots
import os

# Find the directory where scienceplots styles are located
scienceplots_styles_path = os.path.join(os.path.dirname(scienceplots.__file__), 'styles')

# Add the scienceplots styles path to matplotlib's style.core.USER_LIBRARY_PATHS
# This is a less conventional way but can sometimes work when styles aren't
# automatically registered. More robust solution might involve reinstalling
# or checking scienceplots documentation for proper integration.
# However, directly modifying internal paths might have side effects,
# so this is more of a troubleshooting step.

# Check if the path exists and is not already in the list
if os.path.exists(scienceplots_styles_path) and scienceplots_styles_path not in plt.style.core.USER_LIBRARY_PATHS:
    plt.style.core.USER_LIBRARY_PATHS.append(scienceplots_styles_path)
    print(f"Added {scienceplots_styles_path} to matplotlib style paths.")
else:
    print(f"SciencePlots styles path {scienceplots_styles_path} not found or already added.")

# Verify the 'science' style is now available
print("Available matplotlib styles after adding SciencePlots directory:")
print(plt.style.available)
```
